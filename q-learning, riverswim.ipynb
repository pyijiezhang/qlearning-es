```python
import numpy as np
from qlearning import qlearning, qlearning_ucb, qlearning_es, qlearning_es_ucb
```

# env

```python
from riverswim import RiverswimEnv6
env = RiverswimEnv6()
```

```python
env.P 
```

```python
n_steps = 300000
gamma = 0.85
epsilon = 0.1
```

# optimal q

```python
from dp import policy_iteration
```

```python
_, v = policy_iteration(env)
```

```python
q_left = []
for i in range(6):
    if i == 0:
        q = v[i]*gamma+0.05
    else:
        q = v[i-1]*gamma
    q_left.append(q)
```

```python
q_optimal = np.vstack([q_left, v]).T
```

```python
q_optimal
```

# qlearning

```python
Q = qlearning(env, n_steps, gamma, epsilon)
Q
```

```python
for i in [299, 2999, 29999, 299999]:
    q_t = np.array(list(Q[i].values()))
    print(np.abs(q_optimal-q_t).max())
    print()
```

# qlearning_ucb

```python
Q = qlearning_ucb(env, n_steps, gamma, epsilon)
Q
```

```python
for i in [299, 2999, 29999, 299999]:
    q_t = np.array(list(Q[i].values()))
    print(np.abs(q_optimal-q_t).max())
    print()
```
